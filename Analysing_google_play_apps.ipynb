```python
ds['Reviews'] = pd.to_numeric(ds['Reviews'], errors='coerce')
ds
```

```python
ds['Size'].unique()
  
```

```python
def fix_size(s):
  try:
     if s.endswith('M'):
        return float(s[:-1])
     elif s.endswith('k'):
       return float(s[:-1]) / 1024 # convertir en Mb
     elif s == '1,000+':
        return 1.0
     else:
        return np.nan
  except:
      return np.nan
```

```python
fix_size('14M')
```

```python
ds['Size'] = ds['Size'].apply(fix_size)
ds
```

```python
def fix_price(p):
  try:
     if p.startswith('$'):
        return float(p[1:]) # 9otlo ebde khoudh mel position thenya lin yekmel l3dad
     else:
        return 0.0
  except:
      return 0.0
```

```python
ds['Price'] = ds['Price'].apply(fix_price)
ds
```
